### Plotting for Review evaluation and visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('../data/paper.csv')
df['log_size'] = np.log10(df['Dataset_size'])

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Study', y='log_size', hue='Method_family', palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Dataset size (log scale)')
plt.title('Dataset sizes in coin classification papers by method family')
plt.legend(title='Method family')
plt.tight_layout()
plt.savefig('chart1_dataset_sizes.png', dpi=300, bbox_inches='tight')
plt.show()
```

Figure 1. Dataset sizes across reviewed coin ML papers (log scale). 
Classical/statistical methods typically use small datasets (n<1000), 
while CNNs and transformers leverage much larger corpora. 
Data extracted from author-reported numbers; see Table X.

Figure 2. Reported accuracy vs number of classes. High accuracies 
(>90%) are associated with fewer classes and more homogeneous 
datasets; performance drops as class diversity increasesa 
key challenge for ancient coin classification.
